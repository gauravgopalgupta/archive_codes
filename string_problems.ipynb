If you have a string, you have to replace first character with the other one.

```python
s = 'Gaurav'
s = 'S' + s[1:]
```

```python
s = "Gaurav"
s = s.replace('G', 'M', 1)
s
```

## Problem 5 : Occurance of each letter.

#### Finding the count of occurance of each letter in a word.

```python
my_string = 'aabbcccdee'
my_dict = {}

for key in my_string :
  if key not in my_dict.keys() :
    my_dict[key] = 1
  else :
    my_dict[key] += 1

print(my_dict)
```
