## Email Checker

#### program which should split the email and check the correctness of the domain as well as email first part.

```python
import re

def checkmail(email):
    pattern = re.compile("[a-zA-Z0-9]+@[a-z]+")
    #first_part = email.split('.')[0]
    #second_part = email.split('.')[1]
    username = email.split('@')[0]
    web_name = (email.split('@')[1]).split('.')[0]
    extension = email.split('.')[1]
    if (len(extension) != 2) & (len(extension) != 3) :
        return 'invalid last'
    elif pattern.fullmatch(username+'@'+web_name) :
        return 'valid'
    else :
        return 'invalid first'
    #the function should return the strings "invalid" or "valid" based on the email ID entered
email=input()
print(checkmail(email))
```
